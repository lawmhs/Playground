### Normal Distribution:

```python
from scipy.stats import norm
```

```python
pl = figure(title="Normal")
```

```python
x = np.arange(norm.ppf(0.01), norm.ppf(0.99), 0.001)
```

```python
y = norm.pdf(x, 0, 2)
```

```python
source = ColumnDataSource(data={'x': x, 'y': y})
```

```python
pl.line('x', 'y', source=source, line_width=5)
```

```python
show(pl)
```
